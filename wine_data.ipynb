# DATA LOADING

# VISUALIZATIONS

```python
# VIOLINPLOT "PRICE"

plt.figure(figsize=(10,5))
sns.violinplot(data=wine,x=wine["Price"].value_counts(), inner="stick",color="lightgreen")
plt.xlabel("Price IN Euro", fontsize=12)
plt.title("WINE - PRICES - DISTRIBUTION", fontsize=20)
```

```python
# COUNTPLOT "COUNTRIES"
# Sorting ascending with value_counts().index --> descending with value_counts(ascending= True).index

plt.figure(figsize=(8,6))
sns.countplot(data=wine, x="Country", palette="rainbow", order = wine["Country"].value_counts().index)
plt.xlabel("Countries", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("ORIGIN WINE", fontsize=20)

plt.tick_params(axis='both', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.show()
```

```python
# COUNTPLOT "YEAR"

plt.figure(figsize=(10,6))
sns.countplot(data=wine, x="Year", palette="viridis", order = wine["Year"].value_counts().index)
plt.xlabel("Year", fontsize=16)
plt.ylabel("Count", fontsize=16)
plt.title("VINTAGE WINE", fontsize=20)

plt.tick_params(axis='both', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.show()
```

```python
# VIOLINPLOT "WINETYPE" AND "PRICE"

plt.figure(figsize=(7,5))
sns.violinplot(data=wine, x="WineType", y="Price", palette="Set1")
plt.xlabel("Wine Type", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.title("WINE TYPE AND PRICE", fontsize=20)
```

```python
# BOXENPLOT "WINETYPE" AND "PRICE"

plt.figure(figsize=(7,5))
sns.boxenplot(data=wine, x="WineType", y="Price", palette="Set1")
plt.xlabel("Wine Type", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.title("WINE TYPE AND PRICE", fontsize=20)
```

```python
# BOXPLOT "YEAR" AND "PRICE"

plt.figure(figsize=(10,5))
sns.boxplot(data=wine, x="Year", y="Price", palette="coolwarm")
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("VINTAGE AND PRICE", fontsize=20)

plt.tick_params(axis='both', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.show()
```

```python
# HISTPLOT "YEAR" AND "WINETYPE"

plt.figure(figsize=(8,4))
sns.histplot(data=wine, x="Year", hue="WineType", multiple="stack", palette="viridis")
plt.xlabel("Year", )
plt.title("WINE TYPE AND VINTAGE", fontsize=20)

plt.tick_params(axis='both', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.show()
```

```python
# KDE PLOT "RATING" AND "WINETYPE"

sns.kdeplot(data = wine, x = "Rating", hue = "WineType", 
              multiple = "layer", fill = True, alpha = 0.3, linewidth = 1, bw_adjust = 15)
plt.xlabel("Rating", fontsize=16)
plt.ylabel("Distribution", fontsize=16)
plt.title("RATINGS AND WINE TYPE", fontsize=20)
```

```python
# ECDF PLOT "YEAR" AND "WINETYPE"

sns.ecdfplot(data = wine, x = "Rating", hue = "WineType", stat = "count") 
plt.xlabel("Rating", fontsize=16)
plt.ylabel("Distribution", fontsize=16)
plt.title("RATINGS AND WINE TYPE", fontsize=20)
```

```python
# DISPLOT WITH COL(Hist) "RATING" AND "WINETYPE"

sns.displot(data=wine, x="Rating", hue='WineType', multiple="layer", kde=True, col='WineType', bins = 5, 
            palette="coolwarm")
```

```python
# STRIPPLOT "YEAR", "RATING" AND "WINETYPE"
plt.figure(figsize=(10, 5))

sns.stripplot(data=wine, x="Year", y="Rating", hue="WineType", alpha = .9)
plt.xlabel("Year", )
plt.title("WINETYPE - RATING - VINTAGE", fontsize=20)

plt.tick_params(axis='both', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.show()
```
